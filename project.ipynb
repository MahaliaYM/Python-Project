Columna **Age**

```python
import numpy as np

# Ruta del archivo original
#ruta_archivo = 'def/Shark_attacks_cleaned.csv'
ruta_guardado = '/content/drive/MyDrive/Reboot/Labs/Módulo 5.  Python/Proyecto/shar_attack/Shark_attacks_copy_recortado.csv'

# Cargar el dataset
df = pd.read_csv(ruta_guardado)

#Limpieza inicial
def clean_age(age):
    try:
        return int(age)
    except:
        if isinstance(age, str):
            age = age.strip().lower()
            if age == 'teen':
                return (13 + 19) / 2
            elif 's' in age:
                try:
                    decade = int(age.replace('s', ''))
                    return (decade + decade + 9) / 2
                except:
                    return np.nan
        return np.nan

df['Age_clean'] = df['Age'].apply(clean_age)
# Esto es para convertir a enteros pero la columna NO puede tener valores vacíos por lo que se hará al final: df['Age_clean'] = df['Age_clean'].round().astype(int)


# Guardar después de limpieza básica
df.to_csv(ruta_guardado, index=False)
print("CSV guardado tras limpieza básica de 'Age' en la columna Age_clean")
```

```python
# Función auxiliar para imputar edades según la columna 'Name'
def impute_age_from_name(row):
    age = row['Age_clean']
    name = str(row['Victim_Name']).lower() if pd.notnull(row['Victim_Name']) else ""

    if pd.notnull(age):
        return age
    elif "boy" in name:
        return 10
    elif "girl" in name:
        return 10
    elif "child" in name:
        return 8
    elif "baby" in name:
        return 2
    elif "infant" in name:
        return 1
    elif "youth" in name or "youthful" in name:
        return 18
    elif "young man" in name or ("man" in name and "young" in name):
        return 22
    elif "young woman" in name or ("woman" in name and "young" in name):
        return 22
    else:
        return np.nan

# Aplicar la función
df['Age_clean'] = df.apply(impute_age_from_name, axis=1)

# Guardar después de imputación
df.to_csv(ruta_guardado, index=False)
print("CSV guardado tras imputación basada en 'Victim_Name'")
```

```python
# Calcular media de edades
mean_age = df['Age_clean'].mean()

# Rellenar valores NaN con la media
df['Age_clean'] = df['Age_clean'].fillna(round(mean_age)).astype(int)
df['Age_clean'] = df['Age_clean'].round().astype(int)

# Guardar resultado final
df.to_csv(ruta_guardado, index=False)
print(f"CSV guardado con todos los valores completados. Media usada: {round(mean_age)} años.")
```

```python
import pandas as pd
import re
from IPython.display import display

# Ruta del archivo original
#ruta_archivo = 'def/Shark_attacks_cleaned.csv'

# Cargar el dataset
df = pd.read_csv(ruta_guardado)

# ======================
# Normalización de nombres en la columna Country
# ======================
def normalize_country_name(country):
    if pd.isnull(country):
        return country
    country = str(country).strip()
    country = re.sub(r'[^\w\s]', '', country)  # eliminar signos de puntuación
    country = country.title()  # convertir a Título (Asia, United States, etc.)
    corrections = {
        "Usa": "United States",
        "U.S.A": "United States",
        "United States Of America": "United States",
        "Uk": "United Kingdom",
        "Uae": "United Arab Emirates",
        "South Africa?": "South Africa",
        "Asia": "Asia",  # puede incluir más correcciones si conoces errores comunes
    }
    return corrections.get(country, country)

df['Country'] = df['Country'].apply(normalize_country_name)

# ======================
# Función general para imputar por moda de combinación de columnas
# ======================
def impute_with_mode_from_group(df, target_col, key_col1, key_col2):
    mode_map = (
        df[df[target_col].notnull()]
        .groupby([key_col1, key_col2])[target_col]
        .agg(lambda x: x.value_counts().index[0])
    )
    def impute_row(row):
        if pd.isnull(row[target_col]):
            key = (row[key_col1], row[key_col2])
            return mode_map.get(key, row[target_col])
        return row[target_col]
    df[target_col] = df.apply(impute_row, axis=1)
    return df

# ======================
# Guardamos valores faltantes antes
# ======================
cols_to_check = ['Country', 'Region', 'Location']
missing_before = df[cols_to_check].isnull().sum()

# ======================
# Imputaciones por grupos
# ======================
df = impute_with_mode_from_group(df, target_col='Location', key_col1='Country', key_col2='Region')
df = impute_with_mode_from_group(df, target_col='Region', key_col1='Country', key_col2='Location')
df = impute_with_mode_from_group(df, target_col='Country', key_col1='Region', key_col2='Location')

# ======================
# Rellenar lo que queda como 'Unknown'
# ======================
df[cols_to_check] = df[cols_to_check].fillna('Unknown')

# ======================
# Estadísticas después y visualización
# ======================
missing_after = df[cols_to_check].isnull().sum()
changes = pd.DataFrame({
    'Missing Before': missing_before,
    'Missing After': missing_after,
    'Filled': missing_before - missing_after
})

def highlight_filled(val):
    color = 'green' if val > 0 else 'red'
    return f'color: {color}; font-weight: bold'

print("\n=== Imputación completada ===")
display(changes.style.applymap(highlight_filled, subset=['Filled']))

# Guardar archivo actualizado
df.to_csv(ruta_guardado, index=False)
```

Columna **Date**

```python
import pandas as pd
from dateutil import parser
import re

# Cargar el archivo original
df = pd.read_csv(ruta_guardado, encoding='ISO-8859-1')

# Función para extraer y limpiar fechas
def extract_and_parse_date(date_str):
    if pd.isna(date_str) or not isinstance(date_str, str):
        return None, None, None, None

    cleaned = re.sub(r'\b[Rr]eported\b', '', date_str)
    cleaned = cleaned.strip()
    cleaned = re.sub(r'[./]', '-', cleaned)

    try:
        # Detectar y parsear formato ISO (YYYY-MM-DD)
        iso_format = re.match(r'^\d{4}-\d{2}-\d{2}$', cleaned)
        if iso_format:
            parsed = pd.to_datetime(cleaned)
            return parsed.strftime('%d-%m-%Y'), f"{parsed.day:02}", f"{parsed.month:02}", parsed.year

        # Fecha completa en otros formatos
        full_date_match = re.search(r'\d{1,2}-[A-Za-z]{3,9}-\d{4}|\d{1,2}-\d{1,2}-\d{4}', cleaned)
        if full_date_match:
            parsed = parser.parse(full_date_match.group(0), dayfirst=True, fuzzy=True)
            return parsed.strftime('%d-%m-%Y'), f"{parsed.day:02}", f"{parsed.month:02}", parsed.year

        # "Before" + año
        before_year_match = re.search(r'\bBefore\b.*?(\d{4})|\b(\d{4})\s*Before\b', cleaned)
        if before_year_match:
            year = before_year_match.group(1) if before_year_match.group(1) else before_year_match.group(2)
            #return f"01-01-{year}", None, None, int(year)
            return None, None, None, int(year)

        # Palabra + año (como "March 2020")
        word_year_match = re.search(r'\b[A-Za-z]{3,9}\s(\d{4})\b', cleaned)
        if word_year_match:
            parsed = parser.parse(word_year_match.group(0), dayfirst=True, fuzzy=True)
            return parsed.strftime('%d-%m-%Y'), f"{parsed.day:02}", f"{parsed.month:02}", parsed.year

        # Solo año
        year_match = re.search(r'\b(18|19|20)\d{2}\b', cleaned)
        if year_match:
            year = int(year_match.group(0))
            #return f"01-01-{year}", None, None, year
            return None, None, None, year

    except:
        return None, None, None, None

    return None, None, None, None

# Aplicar sin modificar la columna original
df[['new_date', 'days', 'month', 'year']] = df['Date'].apply(
    lambda x: pd.Series(extract_and_parse_date(x))
)

# Guardar archivo limpio
df.to_csv(ruta_guardado, index=False)
```

```python
import pandas as pd
import numpy as np
from keybert import KeyBERT
import gender_guesser.detector as gender

# Cargar dataset
df = pd.read_csv(ruta_guardado) # Poner ruta correcta
df.columns = df.columns.str.strip() #Modifica las columnas para que lea los headings bien

# COLUMNAS NAME Y SEX

# Inicializar modelos
# pip install keybert (y keybert gender-guesser)
kw_model = KeyBERT(model='paraphrase-MiniLM-L6-v2')
detector = gender.Detector(case_sensitive=False)

# Palabras que no son nombres pero pueden dar pistas al keybert
indicadores_masculinos = {"mr", "man", "male", "boy", "gentleman", "sir", "he", "his", "him", "men", "sr"}
indicadores_femeninos = {"mrs", "ms", "miss", "woman", "women", "female", "girl", "lady", "she", "her"}

# Función para extraer una keyword del campo Victim_Name
def extraer_keyword(texto):
    if pd.isna(texto):
        return np.nan
    texto = str(texto).strip().lower()

    if len(texto.split()) <= 2:
        return texto.split()[0]

    keywords = kw_model.extract_keywords(texto, keyphrase_ngram_range=(1, 1), stop_words='english', top_n=1)
    if keywords:
        return keywords[0][0].lower()
    else:
        return np.nan

# Función para deducir sexo desde keyword
def inferir_sexo(row):
    if pd.notna(row["Sex"]):
        return row["Sex"]  # No tocar si ya tiene valor

    keyword = row["Name_keyword"]
    if pd.isna(keyword):
        return np.nan

    # Usar gender-guesser si parece nombre
    guess = detector.get_gender(keyword)

    if guess in ["male", "mostly_male"]:
        return "M"
    elif guess in ["female", "mostly_female"]:
        return "F"

    # Si no es un nombre reconocible, usar pistas
    if keyword in indicadores_masculinos:
        return "M"
    elif keyword in indicadores_femeninos:
        return "F"

    return np.nan

# Extraer keyword
df["Name_keyword"] = df["Victim_Name"].apply(extraer_keyword)

# Aplicar deducción de sexo
df["Sex"] = df.apply(inferir_sexo, axis=1)

# Crear una columna donde los NaN se reemplazan por "Unknown"
df["Sex_unknown"] = df["Sex"].fillna("Unknown")

# Calcular la moda (valor más frecuente)
moda_sexo = df["Sex"].mode().iloc[0]  # Usamos .iloc[0] en caso de empate

# Crear columna nueva rellenando los NaN con la moda
df["Sex_moda"] = df["Sex"].fillna(moda_sexo)

# Normalizar la columna original 'Sex' para modificar valores que no sean "M" o "F"
df["Sex"] = df["Sex"].str.strip().str.upper()  # Elimina espacios y pone en mayúsculas

# Reemplazar valores inválidos por NaN
valores_validos = {"M", "F"}
df["Sex"] = df["Sex"].apply(lambda x: x if x in valores_validos else np.nan)

# Volver a crear las columnas ya limpias
df["Sex_unknown"] = df["Sex"].fillna("Unknown")
moda_sexo = df["Sex"].mode().iloc[0]
df["Sex_moda"] = df["Sex"].fillna(moda_sexo)

# Mostrar un resumen

# Muestra el dataset en el caso de que quieras ver solo los datos de la columna Sex que son NaN
# df_na_sex = df[df["Sex"].isna()]# Filtrar filas donde la columna original Sex es NaN
# print(df_na_sex[["Victim_Name", "Sex", "Sex_unknown", "Sex_moda"]].head(10))

print("Moda del sexo:", moda_sexo) # Muestra cual sexo es el predominante
print(df[["Victim_Name", "Sex", "Sex_unknown", "Sex_moda"]].head(10)) #Muestra del dataset transformado

# Contar valores únicos en Sex_unknown
print("\nValores únicos en 'Sex_unknown':")
print(df["Sex_unknown"].value_counts(dropna=False))  # dropna=False por si quieres ver también NaN (aunque no debería haber)

print("\nValores únicos en 'Sex_moda':")
print(df["Sex_moda"].value_counts(dropna=False))

# Guardar el archivo modificado
df.to_csv(ruta_guardado, index=False)
```

Columna **Species**

```python
import pandas as pd
import re

# Cargar el archivo CSV
df = pd.read_csv(ruta_guardado)

# Lista de frases que indican que no hay especie clara
NEGATIVE_PATTERNS = [
    "no shark", "not confirmed", "highly doubtful",
    "publicity stunt", "doubtful", "not involved", "unconfirmed"
]

# Función para extraer la especie usando regex + filtros negativos
def extract_species_regex(text):
    if not isinstance(text, str):
        return "Unknown"
    text_lower = text.lower()

    # Verificar patrones negativos
    if any(phrase in text_lower for phrase in NEGATIVE_PATTERNS):
        return "Unknown"

    # Verificar si hay "m shark" sin importar mayúsculas/minúsculas y posibles espacios
    if re.search(r'\bm\s*shark\b', text_lower):
        return "Unknown"

    # Buscar especie
    keywords = ['shark', 'hound']
    for keyword in keywords:
        match = re.search(r'\b([a-z\s]+)?(' + keyword + r')\b', text_lower)
        if match:
            return match.group().strip().title()

    return "Unknown"

# Función para extraer longitud
def extract_length(text):
    if not isinstance(text, str):
        return "Unknown"
    text = text.lower().replace(" ", "")
    match_m = re.search(r'(\d+\.?\d*)m', text)
    if match_m:
        return float(match_m.group(1))
    match_ft = re.search(r"(\d+)'", text)
    if match_ft:
        return round(float(match_ft.group(1)) * 0.3048, 2)
    return "Unknown"

# Aplicar funciones
df['species_shark_regex'] = df['Species'].apply(extract_species_regex)
df['long_shark'] = df['Species'].apply(extract_length)

# Guardar el CSV actualizado
df.to_csv(ruta_guardado, index=False)
```

```python
import os
import pandas as pd

ruta_copia = '/content/drive/MyDrive/Reboot/Labs/Módulo 5.  Python/Proyecto/shar_attack/Shark_attacks_copy_recortado_copia.csv'
# Leer el archivo original desde la carpeta 'def'
df = pd.read_csv(ruta_copia, encoding='latin1')

# Eliminar columnas innecesarias
columns_to_drop = [
    "Date", "Year", "Activity", "Sex",
    "Age", "Species", "Name_keyword", "Actividad_Clave"
]
df_cleaned = df.drop(columns=columns_to_drop, errors='ignore')

# Renombrar columnas para mayor claridad
df_cleaned.rename(columns={
    "Age_clean": "Ages",
    "new_date": "Dates",
    "days": "Days",
    "month": "Months",
    "year": "Years",
    "Sex_unknown": "Sex_U",
    "Sex_moda": "Sex_M",
    "Activity_grouped": "Activities",
    "species_shark_regex": "Species_Shark",
    "long_shark ": "Long_Sharks"
}, inplace=True)

# Eliminar filas completamente vacías
df_cleaned.dropna(how='all', inplace=True)

df_cleaned.to_csv(ruta_copia, index=False, encoding='utf-8')

# Mostrar un resumen de los datos limpios
df_cleaned.info()
```
